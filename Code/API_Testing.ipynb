```python
class MPDataDoc:
    def __init__(self, nsites, formula_pretty, material_id, formation_energy_per_atom, fields_not_requested):
        self.nsites = nsites
        self.formula_pretty = formula_pretty
        self.material_id = material_id
        self.formation_energy_per_atom = formation_energy_per_atom
        self.fields_not_requested = fields_not_requested
        
def process_docs(docs):
    summary_docs = []
    for doc in docs:
        nsites = doc['nsites']
        formula_pretty = doc['formula_pretty']
        material_id = doc['material_id']
        formation_energy_per_atom = doc['formation_energy_per_atom']
        fields_not_requested = doc['fields_not_requested']

        summary_doc = MPDataDoc(nsites, formula_pretty, material_id, formation_energy_per_atom, fields_not_requested)
        summary_docs.append(summary_doc)

    return summary_docs

```

```python
def process_documents(docs):
    formula = []
    mpid = []
    formE = []
    
    for doc in docs:
        formula.append(doc.formula_pretty)
        mpid.append(doc.material_id)
        formE.append(doc.formation_energy_per_atom)
        
        
    summary_list = np.column_stack((formula, mpid, formE))
        
    
    return summary_list
```

```python
import csv

def write_csv(data, filename):
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(data)

```
